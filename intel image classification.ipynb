```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob as gb
import cv2 as cv
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D , Dense , MaxPooling2D ,Dropout ,BatchNormalization ,Flatten

trainpath = '../input/intel-image-classification/seg_train/'
testpath = '../input/intel-image-classification/seg_test/'
predpath = '../input/intel-image-classification/seg_pred/'



```

```python



code={'buildings':0,'forest':1,'glacier':2,'mountain':3,'sea':4,'street':5}

def getcode(n):
    for x,y in code.items() :
        if n==y :
            return x
```

```python


s=100

X_train=[]
y_train=[]
X_test=[]
y_test=[]
X_pred=[]

for folder in os.listdir(trainpath+'//seg_train'):
    files=gb.glob(pathname=str(trainpath+'//seg_train//'+folder+'/*.jpg'))
    for file in files :
        image=cv.imread(file)
        X_train.append(list(cv.resize(image,(s,s))))
        y_train.append(code[folder])

# print(len(X_train))





for folder in os.listdir(testpath+'//seg_test'):
    files=gb.glob(pathname=str(testpath+'//seg_test//'+folder+'/*.jpg'))
    for file in files :
        image=cv.imread(file)
        X_test.append(list(cv.resize(image,(s,s))))
        y_test.append(code[folder])


# print(len(X_test))
        
files=gb.glob(pathname=str(predpath+'//seg_pred/*.jpg'))
for file in files :
    image=cv.imread(file)
    X_pred.append(list(cv.resize(image,(s,s))))

# print(len(X_pred))



# plt.figure(figsize=(20,20))
# for n ,i in enumerate(list(np.random.randint(0,len(X_train),36))) :
#     plt.subplot(6,6,n+1)
#     plt.imshow(X_train[i])
#     plt.axis('off')
#     plt.title(getcode(y_train[i]))

X_train=np.array(X_train)
y_train=np.array(y_train)
X_test=np.array(X_test)
y_test=np.array(y_test)
X_pred=np.array(X_pred)





```

```python


KerasModel = Sequential([
        Conv2D(200,kernel_size=(3,3),activation='relu',input_shape=(s,s,3)),
        Conv2D(150,kernel_size=(3,3),activation='relu'),
        MaxPooling2D(4,4),
        Conv2D(120,kernel_size=(3,3),activation='relu'),    
        Conv2D(80,kernel_size=(3,3),activation='relu'),    
        Conv2D(50,kernel_size=(3,3),activation='relu'),
        MaxPooling2D(4,4),
        Flatten() ,    
        Dense(120,activation='relu') ,    
        Dense(100,activation='relu') ,    
        Dense(50,activation='relu') ,        
        Dropout(rate=0.5) ,            
        Dense(6,activation='softmax') ,    
        ])


```

```python
KerasModel.compile(optimizer ='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
ThisModel = KerasModel.fit(X_train, y_train, epochs=50,batch_size=64,verbose=1)

ModelLoss, ModelAccuracy = KerasModel.evaluate(X_test, y_test)
y_pred=KerasModel.predict(X_test)
print(ModelLoss)
print(ModelAccuracy)
```
